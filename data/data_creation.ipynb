```python
import json, random, os

marketplaces = {
    "Amazon": ["AmazonFashion", "AmazonBasics", "PrimeStyle"],
    "eBay": ["TopRatedSeller", "FashionHub", "StyleDeals"],
    "Walmart": ["WalmartFashion", "EverydayStyle", "TrendZone"]
}

brands = ["Hugo Boss", "Armani", "Ralph Lauren", "Calvin Klein", "Tommy Hilfiger"]
categories = {
    "Blazer": (180, 450),
    "Dress Shirt": (60, 150),
    "Formal Pants": (80, 220),
    "Leather Shoes": (120, 350),
    "Tie": (30, 90)
}

def generate_data(marketplace, sellers):
    data = []
    for _ in range(40):
        category = random.choice(list(categories.keys()))
        min_p, max_p = categories[category]
        item = {
            "marketplace": marketplace,
            "title": f"{random.choice(brands)} {category}",
            "category": category,
            "price": round(random.uniform(min_p, max_p), 2),
            "currency": "USD",
            "brand": random.choice(brands),
            "seller": random.choice(sellers),
            "delivery_estimate": f"{random.randint(2,7)} days",
            "url": f"https://www.{marketplace.lower()}.com/item/{random.randint(10000000,99999999)}"
        }
        data.append(item)
    return data

paths = []

for marketplace, sellers in marketplaces.items():
    filename = f"mock_{marketplace.lower()}.json"
    with open(filename, "w") as f:
        json.dump(generate_data(marketplace, sellers), f, indent=2)
    paths.append(filename)

paths
```
